## Conv2D

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
colnames = ['Name', 'FileSize', 'Blocks','Filter', 'Image', 'Output', 'Params', 'J_runtime_h', 'J_frames_h', 'J_runtime_l', 'J_frames_l', 'S_runtime', 'S_frames', 'nan_1', 'nan_2']
data = pd.read_csv('Conv2D.csv', names=colnames, skiprows=1)
data.head()
```

```python
data.plot(kind='line',x='Params',y='FileSize')
plt.show()
```

```python
ax = plt.gca()
data.plot(kind='scatter',x='S_frames',y='S_runtime', color='r', ax=ax)
data.plot(kind='scatter',x='J_frames_l',y='J_runtime_l', color='b', ax=ax)
plt.show()
```
