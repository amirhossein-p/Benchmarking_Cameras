```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
xbar = [1,2,3]
val_r = [397,125,38]
f, ax = plt.subplots(figsize=(10,8))
plt.bar(xbar,val_r)
plt.xticks(xbar, ('JeVois @ 408MHz', 'JeVois @ 1344MHz', 'Sipeed (400MHz)'))
plt.title('MobileNets')
plt.xlabel('Cameras')
plt.ylabel('Runtime [ms]')
plt.show()
```

```python
xbar = [1,2,3]
val_r = [3.3,7.6,26.2]
f, ax = plt.subplots(figsize=(10,8))
plt.bar(xbar,val_r, color='g')
plt.xticks(xbar, ('JeVois @ 408MHz', 'JeVois @ 1344MHz', 'Sipeed (400MHz)'))
plt.title('MobileNets')
plt.xlabel('Cameras')
plt.ylabel('Frames [fps]')
plt.show()
```
