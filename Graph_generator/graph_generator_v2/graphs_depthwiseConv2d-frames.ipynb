## DepthwiseConv2D

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
colnames = ['Name', 'FileSize', 'Blocks', 'Image', 'Output', 'Params', 'J_runtime_h', 'J_frames_h', 'J_runtime_l', 'J_frames_l', 'S_runtime', 'S_frames', 'nan_1', 'nan_2']
data = pd.read_csv('DepthwiseConv2D.csv', names=colnames, skiprows=1)
data.head()
```

```python
ax = plt.gca()
data.plot(kind='scatter',x='Params',y='J_frames_l',c='Image',cmap=plt.cm.get_cmap('plasma'), figsize=(14,10), ax=ax)
plt.title('DepthwiseConv2d - JeVois')
plt.xlabel('Number of parameters')
plt.ylabel('Frames [fps]')
plt.show()
```

```python
ax = plt.gca()
data.plot(kind='scatter',x='Params',y='S_frames',c='Image',cmap=plt.cm.get_cmap('plasma'), figsize=(14,10), ax=ax)
plt.title('DepthwiseConv2d - Sipeed')
plt.xlabel('Number of parameters')
plt.ylabel('Frames [fps]')
plt.show()
```

```python
data.plot(kind='line',x='Params',y='FileSize')
plt.show()
```
